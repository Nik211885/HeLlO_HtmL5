```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
data.head(5)
```

```python
x = data['GrLivArea']
y = data['SalePrice']
x=np.array(x)
y=np.array(y)
x,y
```

```python
thera_0=0
thera_1=0
learning_rate=0.01
```

```python
def draw_line(thera_0,thera_1):
    x_0 = 0
    x_1 = np.max(x)
    y_0=thera_0 + thera_1 *x_0
    y_1=thera_0 + thera_1 *x_1
    plt.plot([x_0,x_1],[y_0,y_1])
```

```python
def paint(thera_0,thera_1):
    draw_line(thera_0,thera_1)
    plt.scatter(x,y,color='red')
    plt.show()
```

```python
paint(thera_0,thera_1)
```

```python
def error(thera_0,thera_1):
    n = x.size
    e=0
    for _ in range(n):
        e += (y[_] - (thera_0 + thera_1 * x[_]))**2
    return e
```

```python
error(thera_0,thera_1)
```

```python
import math
def gradient_descent(thera_0,thera_1):
    n = x.size
    dt_0=0
    x_max = np.max(x)
    y_max = np.max(y)
    a = math.log(y_max,x_max)
    dt_1=0
    for _ in range(n):
        dt_0 += (y[_] - (thera_0 + thera_1 * x[_])) *(-1)
        dt_1 += (y[_] - (thera_0 + thera_1 * x[_])) * (-x[_])
    thera_0 -= dt_0*learning_rate*(1/n**(a*2))
    thera_1 -= dt_1*learning_rate*(1/n**(a*2))
    return thera_0,thera_1
```

```python
a=0
for _ in range(1000):
    thera_0,thera_1=gradient_descent(thera_0,thera_1)
paint(thera_0,thera_1)
```
