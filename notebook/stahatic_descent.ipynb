```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('../data/train.csv')
data.head(10)
```

```python
x = data['GrLivArea']
y = data['SalePrice']
x = np.array(x)
y = np.array(y).reshape(x.size,1)
```

```python
def draw_line(thera_0,thera_1):
    x_0 = 0
    x_1 = np.max(x)
    y_0=thera_0 + thera_1 *x_0
    y_1=thera_0 + thera_1 *x_1
    plt.plot([x_0,x_1],[y_0,y_1])
```

```python
def scater():
    plt.scatter(x,y,color='red')
```

```python
scater()
```

```python
ones = np.ones(x.shape)
temp= np.vstack([ones,x])
a=np.transpose(temp)
a_T = temp
A=np.matmul(a_T,a)
A_T = np.transpose(A)
A_plus = np.matmul(np.linalg.inv(np.matmul(A_T,A)),A_T)
```

```python
b=np.matmul(a_T,y)
```

```python
thera = np.matmul(A_plus,b)
thera_0,thera_1=thera
print(thera_0,thera_1)
```

```python
draw_line(thera_0,thera_1)
scater()
plt.show()
```
